```
# Imports

import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Globals

BASE_URL = "https://statsapi.web.nhl.com"
PLAYER_ID = "8471675" #Sidney Crosby
FIRST_SEASON = "20052006" # Not easily able to figure this out in API yet.
FIRST_SEASON_OF_STATS = "20112012" #NHL kept more stats per game in this year.
```

```
def get_opposite_coord(home, period, period_type, x):
  if home == 1 and period == 2:
    return -1
  elif x < 0:
    return -1
  else:
    return 1

def get_coordinate(play_data, play):

  if(play_data['gameData']['teams']['home']['id'] == TEAM_ID):
    is_home = 1
  else:
    is_home = 0
    

    
  period = play_data['liveData']['plays']['allPlays'][play]['about']['period']
  period_type = play_data['liveData']['plays']['allPlays'][play]['about']['periodType']
    
    
  #print(is_home, period, period_type, play_data['liveData']['plays']['allPlays'][play]['coordinates']['x'], play_data['liveData']['plays']['allPlays'][play]['coordinates']['y'])
  
  play_data = play_data['liveData']['plays']['allPlays'][play]
  get_opposite = get_opposite_coord(is_home, period, period_type, play_data['coordinates']['x'])
  x = play_data['coordinates']['x'] * get_opposite
  y = play_data['coordinates']['y'] * get_opposite
  
  
  return (x, y)

def get_scoring_plays(data):
  return [play_id for play_id in data['liveData']['plays']['scoringPlays']]

def game_parse(player, game_link):
  r = requests.get(BASE_URL + game_link)
  data = r.json()

    
  scoring_plays = get_scoring_plays(data)
  
  for play in scoring_plays:
    play_data = data['liveData']['plays']['allPlays'][play]['players'][0]
    
    if play_data['player']['id'] == int(PLAYER_ID) and play_data['playerType'] == "Scorer":
      GOAL_COORDS.append(get_coordinate(data, play))
```

```
r = requests.get('https://statsapi.web.nhl.com/api/v1/people/' + PLAYER_ID + '/stats?stats=gameLog&season=' + FIRST_SEASON_OF_STATS)
games = r.json()['stats'][0]['splits']

GOAL_COORDS = []

for game in games:
  TEAM_ID = game['team']['id']
  game_parse(PLAYER_ID, game['game']['link'])
```
